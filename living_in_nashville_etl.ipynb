```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Load the data
file_path = ('Park_Locations_cleaned.csv')
df = pd.read_csv(file_path)

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()
```

```python
df.head()
```

```python
# Remove the unwanted columns
df.drop(columns=['Year Established',
        'Picnic Shelters',
        'Picnic Shelters Quantity',
        'Volleyball',
        'Skate Park',
        'Spray Park',
        'Hiking Trails',
        'Horse Trails',
        'Mountain Bike Trails',
        'Boat Launch',
        'Camping Available by Permit',
        'Fishing by Permit',
        'Lake',
        'Canoe Launch',
        'Community Garden',
        'Notes'])
```

```python
df.to_csv('Park_Locations_etl.csv')
```
